```python
df1 =  pd.read_excel('/content/drive/MyDrive/automobile/automobile.xlsx')
df1
```

```python
df2= df1['site_link']
df2
```

```python
url1= df2.values.tolist()
url1
```

```python
len(url1)
```

```python
lis_erreur         =[] 
ville              =[]
code_p             =[]
Mobile             =[]
Adresse            =[]
site_web           =[]

compt=0

try:
  for i in range (len(url1)):
    compt = compt+1
    print(compt, " : ", url1[i])
    req1 =requests.get(url1[i])
    soup1 = bs(req1.text,'html.parser')
    address = soup1.find_all('p',{"class": 'address'})
    city = soup1.find_all('p',{"class": 'city-centre'})
    cp_city = soup1.find_all('p',{"class": 'cp-city'})
    tel  = soup1.find_all('p',{"class": 'tel ico-before'})
    site  = soup1.find_all('div',{"class": 'actions'})

    for i in tel:
        Mobile.append(i.text) #OK
    for i in cp_city:
        code_p.append(i.span.text) #OK
    for i in site:
        site_web .append(i.a.attrs['href'])#ok
    for i in address:
        Adresse.append(i.text) #OK
    for i in city :
        ville .append(i.text)
except:
      print(compt,':  ',url1[i] )
      lis_erreur.append(url1[i])
```

```python
df_erreur= pd.DataFrame(lis_erreur)
df1= pd.DataFrame(Mobile)
df2= pd.DataFrame(Adresse)
df3= pd.DataFrame(site_web)
df4= pd.DataFrame(code_p)
df = pd.concat([df1,df2,df3,df4], axis=1)
df_erreur.to_excel("/content/drive/MyDrive/automobile/df_erreur1.xlsx")
df.to_excel("/content/drive/MyDrive/automobile/AUTO1.xlsx")
df
```

```python
prenom             =[]
Nom                =[]
lis_erreur         =[]
Email              =[]
ville              =[]
code_p             =[]
Mobile             =[]
Metier             =[]
Adresse            =[]
site_web           =[]
Activites          =[]
type_professionnel =[]
compt=0
for g in range (len(url1)):
    #print(url1[g])
    req1 =requests.get(url1[g])
    soup1 = bs(req1.text,'html.parser')
    s_nom_prenom = soup1.find_all('td',{"class": 'views-field views-field-title'})
    s_type_professionnel=soup1.find_all('td',{"class": 'views-field views-field-field-ads-annuaire-categorie'})
    s_adresse = soup1.find_all('div',{"class": 'street-block'})
    s_code_Postal_ville = soup1.find_all('div',{"class": 'addressfield-container-inline locality-block country-FR'})
    s_prenom = soup1.find_all('td',{"class": 'views-field views-field-field-ads-annuaire-prenom'})
    url_p = soup1.find_all('td',{"class": 'views-field views-field-title'})
    s_code_Postal = soup1.find_all('div',{"class": 'addressfield-container-inline locality-block country-FR'})

    for elt in url_p:
        compt= compt+1
        print(compt," https://www.inserr.fr/"+elt.a.attrs['href'])
        site_web.append("https://www.inserr.fr/"+elt.a.attrs['href'])
        try:
            req2 =requests.get("https://www.inserr.fr/"+elt.a.attrs['href'])
            soup2 = bs(req2.text,'html.parser')
            all_data = soup2.find_all('div',{"class": 'field-item even'})
            Email.append(all_data[0].text)
            Mobile.append(all_data[1].text)
            Activites.append(all_data[3].text)
        except:
              print(" erreur site: ",g,": ", "https://www.inserr.fr/"+elt.a.attrs['href'])
              lis_erreur.append("https://www.inserr.fr/"+elt.a.attrs['href'])
              #print("erreur : ", elt)
    for i in s_adresse:
        Adresse.append(i.text)
    for j in s_nom_prenom:
        Nom.append(j.a.text)
    
    for k in s_prenom:
        prenom.append(k.text)

    for l in s_type_professionnel:
        type_professionnel.append(l.text)

    for m in s_code_Postal_ville:
        ville.append(m.text)

    for n in s_code_Postal:
        code_p.append(n.span.text)

df_erreur= pd.DataFrame(lis_erreur)
df1= pd.DataFrame(Email)
df2= pd.DataFrame(Mobile)
df3= pd.DataFrame(Adresse)
df4= pd.DataFrame(Activites)
df5= pd.DataFrame(site_web)
df6= pd.DataFrame(Nom)
df7= pd.DataFrame(type_professionnel)
df8= pd.DataFrame(prenom)
df9= pd.DataFrame(code_p)
df = pd.concat([df1,df2,df3,df4,df5,df6,df7,df8,df9], axis=1)
df_erreur.to_excel("/content/drive/MyDrive/Inserr/df_erreur.xlsx")
df.to_excel("/content/drive/MyDrive/Inserr/Inserr3.xlsx")
df
```
